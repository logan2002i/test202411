```python
import os
import cv2
#from mtcnn import MTCNN
from mtcnn_cv2 import MTCNN
detector = MTCNN()
```

```python
def extract_file(path):
    #path 圖片資料夾絕對路徑
    pach_list = []
    if os.path.isdir(path) == True:
            for r, d, f in os.walk(path): # r=root, d=directories, f = files
                for file in f:
                    if ('.jpg' in file):
                        exact_path = r + "/" + file
                        pach_list.append(exact_path)
#                         print(exact_path)
                    else:
                        print("沒照片")
    return pach_list
```

```python
def mtcnn_process(img_path, save_path,file):
    '''
    #MTCNN bounding box 說明
    #回傳四個參數   x1, y1, w, h
    # x1, y1為起始座標 , w為x軸長度 , h為y軸長
    opencv 裁切為[y起始:y結束, x起始:x結束]
    '''
    # img_path 圖片絕對路徑
    # save_path 放圖片資料夾路徑
    # years 年齡
    # file 圖檔名稱
    detector = MTCNN()
    test_pic = img_path
    image = cv2.cvtColor(cv2.imread(test_pic), cv2.COLOR_BGR2RGB)
    result = detector.detect_faces(image)
#     print(img_path)
#     print(result)
    if len(result)<0:
        print("偵測不到傻眼")
        print(img_path+"偵測不到傻眼")
        print(result[0]['box'])
    if len(result) > 0:
        bounding_box = result[0]['box']

        crop_img = image[bounding_box[1]:bounding_box[1]+bounding_box[3], bounding_box[0]:bounding_box[0]+bounding_box[2]]
        resize_img = cv2.resize(crop_img,(112,112))
#         if years == None:
        cv2.imwrite(save_path+'/'+file, cv2.cvtColor(resize_img, cv2.COLOR_RGB2BGR))
#         else:
#         cv2.imwrite(save_path+'/'+str(years)+'_'+file, cv2.cvtColor(resize_img, cv2.COLOR_RGB2BGR))
```
